```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
pd.set_option("display.max_columns", 500)
matplotlib.rcParams["figure.figsize"]=(20,10)
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
case_time_series_url='https://data.covid19india.org/csv/latest/case_time_series.csv' 
states_url='https://data.covid19india.org/csv/latest/states.csv' 
districts_url='https://data.covid19india.org/csv/latest/districts.csv'
state_wise_daily_url='https://data.covid19india.org/csv/latest/state_wise_daily.csv'
state_wise_url='https://data.covid19india.org/csv/latest/state_wise.csv'
district_wise_url='https://data.covid19india.org/csv/latest/district_wise.csv'
statewise_tested_numbers_data_url='https://data.covid19india.org/csv/latest/statewise_tested_numbers_data.csv'
tested_numbers_icmr_data_url=	'https://data.covid19india.org/csv/latest/tested_numbers_icmr_data.csv'
icmr_labs_statewise_url=	'https://data.covid19india.org/csv/latest/icmr_labs_statewise.csv'
sources_list_url=	'https://data.covid19india.org/csv/latest/sources_list.csv'
rtpcr_samples_collected_url=	'http://data.covid19india.org/csv/latest/icmr_rtpcr_tests_daily.csv'
vaccine_doses_administered_statewise_url=	'http://data.covid19india.org/csv/latest/vaccine_doses_statewise_v2.csv'
cowin_vaccine_data_statewise_url=	'http://data.covid19india.org/csv/latest/cowin_vaccine_data_statewise.csv' 
cowin_vaccine_data_districtwise_url=	'http://data.covid19india.org/csv/latest/cowin_vaccine_data_districtwise.csv'
```

```python
#function to read the csv files
def read_data(url):
  df=pd.read_csv(url)
  return df
```

```python
case_time_series=read_data(case_time_series_url)
states=read_data(states_url)
districts = read_data(districts_url)
state_wise_daily = read_data(state_wise_daily_url)
district_wise = read_data(district_wise_url)
statewise_tested_numbers_data = read_data(statewise_tested_numbers_data_url)
tested_numbers_icmr_data = read_data(tested_numbers_icmr_data_url)
icmr_labs_statewise = read_data(icmr_labs_statewise_url)
sources_list = read_data(sources_list_url)
rtpcr_samples_collected = read_data(rtpcr_samples_collected_url)
vaccine_doses_administered_statewise = read_data(vaccine_doses_administered_statewise_url)
cowin_vaccine_data_statewise = read_data(cowin_vaccine_data_statewise_url)
cowin_vaccine_data_districtwise = read_data(cowin_vaccine_data_districtwise_url)
```

```python
states['Active'] = states['Confirmed'] - states['Recovered']
```

```python
states
```

```python
states['Date'] = pd.to_datetime(states['Date'])
states['Date'] = states['Date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
states_pop = states.groupby(['Date','State']).agg({'Confirmed':'sum','Recovered':'sum','Deceased':'sum','Active':'sum'}).reset_index()
states_pop
```

```python
states_pop['Recovered%'] = (states_pop['Recovered']/states_pop['Confirmed'])*100
states_pop['Deceased%'] = (states_pop['Deceased']/states_pop['Confirmed'])*100
states_pop['Active%'] = (states_pop['Active']/states_pop['Confirmed'])*100
states_pop
```

```python
districts
```

```python
districts['Date'] = pd.to_datetime(districts['Date'])
districts['Date'] = districts['Date'].apply(lambda x: x.strftime('%Y-%m'))

districts['Active'] = districts['Confirmed'] - districts['Recovered']

districts_pop_state = districts.groupby(['Date','State']).agg({'Confirmed':'sum','Recovered':'sum','Deceased':'sum','Active':'sum'}).reset_index()

districts_pop_state['Recovered%'] = (districts_pop_state['Recovered']/districts_pop_state['Confirmed'])*100
districts_pop_state['Deceased%'] = (districts_pop_state['Deceased']/districts_pop_state['Confirmed'])*100
districts_pop_state['Active%'] = (districts_pop_state['Active']/districts_pop_state['Confirmed'])*100

districts_pop_districs = districts.groupby(['Date','State','District']).agg({'Confirmed':'sum','Recovered':'sum','Deceased':'sum','Active':'sum'}).reset_index()

districts_pop_districs['Recovered%'] = (districts_pop_districs['Recovered']/districts_pop_districs['Confirmed'])*100
districts_pop_districs['Deceased%'] = (districts_pop_districs['Deceased']/districts_pop_districs['Confirmed'])*100
districts_pop_districs['Active%'] = (districts_pop_districs['Active']/districts_pop_districs['Confirmed'])*100
```

```python
statewise_tested_numbers_data['Updated On'] = pd.to_datetime(statewise_tested_numbers_data['Updated On'])
statewise_tested_numbers_data['Updated On'] = statewise_tested_numbers_data['Updated On'].apply(lambda x: x.strftime('%Y-%m'))

statewise_tested_numbers_data_pop = statewise_tested_numbers_data.groupby(['Updated On','State']).agg({'Total Tested':'sum'}).reset_index()
```

```python
statewise_tested_numbers_data_pop
```

```python
vaccine_doses_administered_statewise
```

```python
vaccine_doses_administered_statewise['Vaccinated As of'] = pd.to_datetime(vaccine_doses_administered_statewise['Vaccinated As of'])
vaccine_doses_administered_statewise['Vaccinated As of'] = vaccine_doses_administered_statewise['Vaccinated As of'].apply(lambda x: x.strftime('%Y-%m'))

vaccine_doses_administered_statewise_pop = vaccine_doses_administered_statewise.groupby(['Vaccinated As of','State']).agg({'First Dose Administered':'sum','Second Dose Administered':'sum',
                                                                                                                           'Total Doses Administered':'sum'}).reset_index()
```

```python
vaccine_doses_administered_statewise_pop
```

```python
cowin_vaccine_data_statewise
```

```python
cowin_vaccine_data_statewise['Updated On'] = pd.to_datetime(cowin_vaccine_data_statewise['Updated On'])
cowin_vaccine_data_statewise['Updated On'] = cowin_vaccine_data_statewise['Updated On'].apply(lambda x: x.strftime('%Y-%m'))

cowin_vaccine_data_statewise_pop = cowin_vaccine_data_statewise.groupby(['Updated On','State']).agg({'Sessions':'sum','Sites':'sum','First Dose Administered':'sum','Second Dose Administered':'sum',
                                                                                                                           'Total Doses Administered':'sum','Covaxin (Doses Administered)':'sum','CoviShield (Doses Administered)':'sum'}).reset_index()
```

```python
cowin_vaccine_data_statewise_pop['Covaxin (Doses Administered)%'] = (cowin_vaccine_data_statewise_pop['Covaxin (Doses Administered)']/cowin_vaccine_data_statewise_pop['Total Doses Administered'])*100
cowin_vaccine_data_statewise_pop['CoviShield (Doses Administered)%'] = (cowin_vaccine_data_statewise_pop['CoviShield (Doses Administered)']/cowin_vaccine_data_statewise_pop['Total Doses Administered'])*100
```
